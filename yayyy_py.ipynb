```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/CAR DETAILS FROM CAR DEKHO.csv")

df["name_2"] = df["name"].apply(lambda x: x.split(" ")[0])

labels = df["name_2"][:30].value_counts().index
sizes = df["name_2"][:30].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=45)
plt.title('name', color='red', fontsize=15)
plt.show()
```
